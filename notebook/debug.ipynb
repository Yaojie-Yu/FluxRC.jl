```julia
using Kinetic, FastGaussQuadrature
```

```julia
x0 = -1; x1 = 1
ncell = 100
nface = ncell + 1
dx = (x1 - x0) / ncell

p = 2
nsp = p + 1

cfl = 0.2
```

```julia
xgl = gausslegendre(nsp)[1]
```

```julia
x = collect(x0:dx:x1);
```

```julia
xin = zeros(ncell, nsp)
u = zeros(ncell, nsp)
for i=1:ncell, ppp1=1:nsp
    xin[i,ppp1]=((1-xgl[ppp1])/2.0)*x[i]+((1+xgl[ppp1])/2.0)*x[i+1]
    u[i,ppp1]=exp(-20.0*xin[i,ppp1]^2)
end
uold = deepcopy(u);
```

```julia
function lagrange_poly!(l, xi, xgl)
    for k=1:nsp
        term=1.0

        for j=1:nsp
            if j != k
                term *= (xi-xgl[j]) / (xgl[k]-xgl[j])
            end
        end
        l[k]=term        
    end
end
```

```julia
ll = zeros(nsp)
lr = zeros(nsp)
```

```julia
lagrange_poly!(ll, -1.0, xgl)
lagrange_poly!(lr, 1.0, xgl)
```

```julia
f = zeros(ncell, nsp);
u_face = zeros(ncell, 2)
f_face = zeros(ncell, 2);
```

```julia
t = 0
```

```julia
for iter in 1:200
    
    for i in 1:ncell, j in 1:nsp
        f[i,j] = uold[i,j]
    end

    u_face .= 0.
    f_face .= 0.
    for i=1:ncell, j in 1:nsp
        # right face of element i
        u_face[i,1]+=uold[i,j]*lr[j]
        f_face[i,1]+=f[i,j]*lr[j]

        # left face of element i
        u_face[i,2]+=uold[i,j]*ll[j]
        f_face[i,2]+=f[i,j]*ll[j]    
    end

    au = zeros(nface);
    au[1] = (f_face[1,2]-f_face[ncell,1])/(u_face[1,2]-u_face[ncell,1]+1e-6)
    au[nface] = (f_face[1,2]-f_face[ncell,1])/(u_face[1,2]-u_face[ncell,1]+1e-6)
    for i=2:nface-1
        au[i] = (f_face[i,2]-f_face[i-1,1])/(u_face[i,2]-u_face[i-1,1]+1e-6)
    end

    f_interaction = zeros(nface)
    f_interaction[1]=(0.5*(f_face[ncell,1]+f_face[1,2])-0.5*abs(au[1])*(u_face[1,2]-u_face[ncell,1]))
    f_interaction[nface]=(0.5*(f_face[ncell,1]+f_face[1,2])-0.5*abs(au[nface])*(u_face[1,2]-u_face[ncell,1]))
    for i=2:nface-1
        f_interaction[i]=(0.5*(f_face[i-1,1]+f_face[i,2])-0.5*abs(au[i])*(u_face[i,2]-u_face[i-1,1]))
    end

    lpdm = zeros(nsp, nsp)
    for k=1:nsp, m=1:nsp
        lsum = 0.
        for l=1:nsp
            term = 1.
            for j=1:nsp
                if j!=k && j!=l
                    term *= (xgl[m]-xgl[j]) / (xgl[k] - xgl[j])
                end
            end
            if l!=k
              lsum += term / (xgl[k] - xgl[l])
            end
        end
        lpdm[m,k] = lsum
    end

    dlpdr = 3. .* xgl
    dlpp1dr = 15 / 2 .* xgl.^2 .- 1.5
    dgl = (-1.)^2 * 0.5 * (dlpdr .- dlpp1dr)
    dgr = 0.5 .* (dlpdr .+ dlpp1dr)

    rhs1 = zeros(ncell, nsp)
    df = zeros(ncell, nsp)

    # Vector-matrix multiplication (entire first term on RHS)
    for i=1:ncell, ppp1=1:nsp, k in 1:nsp
        rhs1[i,ppp1]+=f[i,k]*lpdm[ppp1,k]
    end

    # Computing RHS,i.e., divergence of flux
    for i=1:ncell, ppp1=1:nsp
        df[i,ppp1] = -(rhs1[i,ppp1] + (f_interaction[e2f[i,2]]-f_face[i,2])*dgl[ppp1] + (f_interaction[e2f[i,1]]-f_face[i,1])*dgr[ppp1])
    end

    k1 = dt * df
    @. uold = u + k1
    u .= uold
    
    t+=dt
end
```

```julia
au[1] = (f_face[1,2]-f_face[ncell,1])/(u_face[1,2]-u_face[ncell,1]+1e-6)
au[nface] = (f_face[1,2]-f_face[ncell,1])/(u_face[1,2]-u_face[ncell,1]+1e-6)
for i=2:nface-1
    au[i] = (f_face[i,2]-f_face[i-1,1])/(u_face[i,2]-u_face[i-1,1]+1e-6)
end
```

```julia
f_interaction = zeros(nface)

f_interaction[1]=(0.5*(f_face[ncell,1]+f_face[1,2])-0.5*abs(au[1])*(u_face[1,2]-u_face[ncell,1]))
f_interaction[nface]=(0.5*(f_face[ncell,1]+f_face[1,2])-0.5*abs(au[nface])*(u_face[1,2]-u_face[ncell,1]))
for i=2:nface-1
    f_interaction[i]=(0.5*(f_face[i-1,1]+f_face[i,2])-0.5*abs(au[i])*(u_face[i,2]-u_face[i-1,1]))
end
```

```julia
lpdm = zeros(nsp, nsp)
for k=1:nsp, m=1:nsp
    lsum = 0.
    for l=1:nsp
        term = 1.
        for j=1:nsp
            if j!=k && j!=l
                term *= (xgl[m]-xgl[j]) / (xgl[k] - xgl[j])
            end
        end
        if l!=k
          lsum += term / (xgl[k] - xgl[l])
        end
    end
    lpdm[m,k] = lsum
end
```

```julia
dlpdr = 3. .* xgl
dlpp1dr = 15 / 2 .* xgl.^2 .- 1.5
```

```julia
dgl = (-1.)^2 * 0.5 * (dlpdr .- dlpp1dr)
dgr = 0.5 .* (dlpdr .+ dlpp1dr)
```

```julia
e2f = zeros(Int, ncell, 2)
for i=1:ncell
    if i==1
        e2f[i,2]=nface
        e2f[i,1]=i+1
    elseif i==ncell
        e2f[i,2]=i
        e2f[i,1]=1
    else
        e2f[i,2]=i
        e2f[i,1]=i+1
    end
end
```

```julia
rhs1 = zeros(ncell, nsp)
df = zeros(ncell, nsp)

# Vector-matrix multiplication (entire first term on RHS)
for i=1:ncell, ppp1=1:nsp, k in 1:nsp
    rhs1[i,ppp1]+=f[i,k]*lpdm[ppp1,k]
end

# Computing RHS,i.e., divergence of flux
for i=1:ncell, ppp1=1:nsp
    df[i,ppp1] = -(rhs1[i,ppp1] + (f_interaction[e2f[i,2]]-f_face[i,2])*dgl[ppp1] + (f_interaction[e2f[i,1]]-f_face[i,1])*dgr[ppp1])
end
```

```julia
dt = (cfl*dx)
```

```julia
k1=dt*df
@. uold = u + k1
```
